```python
def get_text(dir):
    out = ""
    with open(dir, "r") as f:
        lines = f.readlines()
        for line in lines:
            out+=line
    return out
```

```python
import re

def get_interventions(text):
    regex = "El señor [A-Z][A-Z]+|La señora [A-Z][A-Z]+"
    out = []

    for m in re.finditer(regex, text):
        l = m.span()[0]
        r = m.span()[1]
        parentesis = 0

        while r + 1 < len(text):
            if text[r] == ":":
                break

            if text[r] == "(":
                parentesis = 1
            if text[r] == ")":
                parentesis = 0

            if re.match("[a-z]", text[r]) and not parentesis:
                r -= 1
                break

            r+=1
        
        out.append([text[l:r], (l, r)])

    for i in range(len(out)-1):
        out[i][1] = text[out[i][1][1]+1:out[i+1][1][0]].strip()
    out.pop()
    return out
    
```

```python
import os

dir1 = r""
dir2 = r""

for name in os.listdir(dir1):

    # LEER EL NÚMERO DE CADA DIARIO Y SU LEGISLATURA

    tmp = name[3:]
    legis = ""
    for char in tmp:
        if char != "_":
            legis+=char
        else:
            break
    # print(legis)
    num = ""
    tmp = tmp[len(legis)+1:]
    for char in tmp:
        if char != ".":
            num+=char
        else:
            break
    # print(num)
    legis = int(legis)
    num = int(num)

    try:
        text = get_text(os.path.join(dir1, name))
        with open(os.path.join(dir2, name), "w") as f:
            for [interventor, intervention] in get_interventions(text):
                f.write(interventor)
                f.write("\n")
                f.write(intervention)
                f.write("\n")
    except:
        print(legis)
        print(num)
```
